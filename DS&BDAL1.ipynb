```python
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
print("Average of stroke:", avg_stroke)
```

```python
df["stroke"].replace(np.nan, avg_stroke, inplace = True)
```

```python
avg_hp = df["horsepower"].astype("float").mean(axis = 0)
print("Average of horsepower:", avg_stroke)
```

```python
df["horsepower"].replace(np.nan, avg_hp, inplace = True)
```

```python
avg_prpm = df["peak-rpm"].astype("float").mean(axis = 0)
print("Average of peak-rpm:", avg_prpm)
```

```python
df["peak-rpm"].replace(np.nan, avg_prpm, inplace = True)
```

```python
mode_hpb = df["horsepower-binned"].astype("category").mode().iloc[0]
print("Mode of horsepower-binned:", mode_hpb)
```

```python
df["horsepower-binned"].replace(np.nan, mode_hpb, inplace = True)
```
